```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('russian')) 

def remove_stopwords(text):
    tokens = text.split()  
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words] 
    return ' '.join(filtered_tokens) 

df_norm['description'] = df_norm['description'].apply(remove_stopwords)
df_norm['about'] = df_norm['about'].apply(remove_stopwords)
df_norm
```

```python
import re

def remove_punctuation(text):
    cleaned_text = re.sub(r'[^\w\s]', ' ', text)
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text)
    return cleaned_text.strip()

df_norm['description'] = df_norm['description'].apply(remove_punctuation)
df_norm['about'] = df_norm['about'].apply(remove_punctuation)
df_norm['key_skills'] = df_norm['key_skills'].apply(remove_punctuation)
df_norm
```

```python
from datetime import datetime

def calculate_unix_diff(row):
    starts = row['experienceItem'][0]['starts']
    ends = row['experienceItem'][0]['ends']
    
    start_unix = pd.to_datetime(starts).timestamp()
    
    if ends:
        end_unix = pd.to_datetime(ends).timestamp()
    else:
        end_unix = datetime.now().timestamp()

    return end_unix - start_unix

df_new['experience_unix_diff'] = df_new.apply(calculate_unix_diff, axis=1)
df_new = df_new.drop('experienceItem', axis=1)
```
